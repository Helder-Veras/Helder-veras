### Pegar 1 coluna ( e os pd.Series)

### Utilização do método loc
    -Pegar uma linha
    -Pegar linhas de acordo com uma condição específica
    -Pegar linahs e colunas específicas
    -Pegar um valor específico

```python
# pegar uma linha específica
display(vendas.loc[1:5])

# pegar linhas que correspondem a uma condição
display(vendas.loc[vendas['Produto']=='Camiseta'])

#pegar várias linhas e colunas usando o loc
vendas_northshopping = vendas.loc[vendas['ID Loja']=='Norte Shopping',["ID Loja","Produto","Quantidade"]]
display(vendas_northshopping)
```

### Adicionar 1 coluna

```python
# a partir de uma coluna que já existe
vendas['Comissão'] = vendas['Valor Final']*0.05

# criar uma coluna com valor padrão
vendas.loc[:,'Imposto'] = 0
display(vendas)
```

### Adicionar 1 linha

```python
vendas_dez = pd.read_excel(r'C:\Users\helde\Downloads\Vendas - Dez.xlsx')

#adicionar
vendas_final = vendas.append(vendas_dez)
display(vendas_final)
```

### Excluir linhas e colunas 

```python
vendas_final2 = vendas_final.drop("Imposto",axis = 1)   # axis = 0 é a linha, axis = 1 é a coluna
display(vendas_final2)
```

### Valores vazios
     -Deletar linhas/colunas vazias
     -Deletar linahs que possuem valores vazios
     -Preencher os valores vazios ( média e último valor)

```python
# deletar linhas e colunas completamente vazias 
vendas_final3 = vendas_final2.dropna(how = 'all',axis = 1)

# deletar completamente a linha que possuem pelo menos 1 valor vazio

vendas_final4 = vendas_final2.dropna()

```
